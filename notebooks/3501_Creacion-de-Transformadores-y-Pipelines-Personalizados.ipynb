## Imports

```python
import arff
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
```

```python
def load_kdd_dataset(data_path):
    """Lectura del DataSet NLS-KDD"""
    with open(data_path, 'r')as train_set:
        dataset = arff.load(train_set)
    attributes =[attr[0]for attr in dataset['attributes']]
    return pd.DataFrame(dataset['data'],columns = attributes)
```

```python
def train_val_test_split(df, rstate=42, shuffle=True, stratify=None):
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

# API Sklearn

Antes de continuar vamos a hacer una reseña sobre como funcionan las Api de Sklearn:
* **Estimators:** Cualquier objeto que pueda estimar algún parametro:
    * El propio estimador se forma mediante el método fit(), que siempre toma un DataSet, como argumento.
    * Cualquier otro parámetro de este método, es un hiperparámetro.
* **Transformer**: Son estimadores capaces de transformar el DataSet(como inputer).
    * La transformación se realiza mediante el método transform().
    * Reciben un DataSet como parámetro de entrada.
* **Predictors**: Son estimadores capaces de realizar predicciones.
    * La predicción se realiza mediante el método predict().
    * Reciben un DataSet como entrada.
    * RETORNAN UN DataSet con las predicciones.
    * Tienen un método score() para evaluar el resultado de la predicción

### Transformadores para atributos numéricos

```python
from sklearn.base import BaseEstimator, TransformerMixin

class DeleteNanRows(BaseEstimator, TransformerMixin):
    def _init_self():
        pass
    def fit(self, X, y= None):
        return self
    def transform(self, X, y= None):
        return X.dropna()
```

```python
delete_nan = DeleteNanRows()
X_train_prep = delete_nan.fit_transform(X_train)
```

```python
X_train_prep
```

```python
# Transformador diseñado para escalar de manera sencilla únicamente
# algunas columnas seleccionadas.
class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, attributes):
        self.attributes = attributes
    def fit(self, X, y = None):
        return self # Nothing else to do
    def transform(self, X, y = None):
        X_copy = X.copy()
        scale_attr = X_copy[self.attributes]
        robust_scaler = RobustScaler() # Igualar los datos para que todos tengan la misma escala, reduce el impacto de los valores
        X_scaled = robust_scaler.fit_transform(scale_attr)
        X_scaled = pd.DataFrame(X_scaled, columns=self.attributes, index=X_copy.index)
        for attr in self.attributes:
            X_copy[attr] = X_scaled[attr]
        return X_copy
```

## 6.- Construyendo Pipelines Personalizados
Los Pipelines permiten agrupar en un flujo de ejecución todas las operaciones de transformación que se necesitan realizar para sobre el DataSet, esto facilita muchisisimo las transformaciones para diferentes DataSets.
Los aspectos a tener en cuenta de estas estructuras son:
* Recibe un DataSet de pares (nombre, estimador)
* Todos menos el último deben ser transformadores.
* El Pipeline expone los mismos métodos que el **último estimador**.
* Cuando se llama el método fit() del Pipeline, se llama secuencialmente al método fit_transform() de los estimadores y se les pasa de manera secuencial el output del siguiente. El último invoca al método fit()

```python
# Construcción de un Pipeline para los atributos numéricos.
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('rbst_scaler', RobustScaler()),
])
```

```python
# La clase imputer no admite valores categoricos,
# se eliminan los atributos categoricos
X_train_num = X_train.select_dtypes(exclude=['object'])

X_train_prep = num_pipeline.fit_transform(X_train_num)
X_train_prep = pd.DataFrame(X_train_prep, columns=X_train_num.columns, index=X_train_num.index)
```

A continuación se presenta el método **ColumnTransformer**, que ejecuta todos los pipelines y concatena el resultado para ello
el resultado de los Pipelines debe ser un valor numerico.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

num_attribs = list(X_train.select_dtypes(exclude = ['object']))
cat_attribs = list(X_train.select_dtypes(include = ['object']))

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])
```

```python
X_train_prep = full_pipeline.fit_transform(X_train)
```

```python
X_train_prep = pd.DataFrame(X_train_prep, columns=list(pd.get_dummies(X_train)), index= X_train.index)
```
